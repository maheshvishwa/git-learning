```
##Select the latest active subscriptions for each account###
active_list = spark.sql(f"SELECT shaw_account_no, shaw_subscriber_no,RANK() OVER(ORDER BY create_date DESC) RNK FROM file_seq_ban_sub_event where (load_type = 'LD' and status_code = 'C')")
active_list = active_list.select("shaw_account_no","shaw_subscriber_no").filter(active_list.RNK == "1")
baninfo = spark.sql("select * from shaw_baninfo_migr where ban_status <> 'O'").withColumnRenamed("drvd_insert_ts","ts1")
subinfo = spark.sql("select * from shaw_subinfo_migr where sub_status = 'I'").withColumnRenamed("drvd_insert_ts","ts2")
```
